```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import load_iris
```

```python
datsets=load_iris()
```

```python
print(datsets.DESCR)
```

```python
datsets.feature_names
```

```python
datsets.target_names
```

```python
datsets.target
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=datsets.target
```

```python
x
```

```python
y
```

### Note : We dont need to perform feature scaling in Decision tree algorithms 

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=90)
```

### Modelling using port purning 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree 
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

```python
#post pruning 
classifier_1=DecisionTreeClassifier(criterion='entropy',max_depth=2)
```

```python
classifier_1.fit(x_train,y_train)
```

```python
from sklearn import tree 
plt.figure(figsize=(12,10))
tree.plot_tree(classifier_1,filled=True)
```

```python
y_pred=classifier_1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('Accuracy:',accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Prepruning using Hyperperameter tuning 

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier_2=DecisionTreeClassifier()
clt=GridSearchCV(classifier_2,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clt.fit(x_train,y_train)
```

```python
clt.best_params_
```

```python
clt.best_score_
```

```python
y_pred=clt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('Accuracy:',accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Decision Tree Regreessor

```python
from sklearn.datasets import fetch_california_housing
```

```python
dataset=fetch_california_housing()
```

```python
print(dataset.DESCR)
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df['target']=dataset.target
```

```python
df.head()
```

```python
# taking a sample of data 
df=df.sample(frac=0.25)
```

```python
df.shape
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=78)
```

```python
from sklearn.tree import DecisionTreeRegressor

```

```python
Mode_reg=DecisionTreeRegressor()
```

```python
Mode_reg.fit(x_train,y_train)
```

```python
y_pred=Mode_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
## Hyperparameter Tunning
parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
  'max_features':['auto', 'sqrt', 'log2']
    
}
regressor=DecisionTreeRegressor()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
regressorcv=GridSearchCV(regressor,param_grid=parameter,cv=2,scoring='neg_mean_squared_error')
```

```python
regressorcv.fit(x_train,y_train)
```

```python
regressorcv.best_params_
```

```python
regressorcv.best_score_
```

```python
y_pred_2=regressorcv.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_2)
```

```python
ds_model=DecisionTreeRegressor(criterion='absolute_error',max_depth=7,splitter='best')
```

```python
ds_model.fit(x_train,y_train)
```

```python
%matplotlib inline
from sklearn import tree 
plt.figure(figsize=(12,10))
tree.plot_tree(ds_model,filled=True)
plt.show()
```

```python
y_pred_3=ds_model.predict(x_test)
```

```python
print('Model Accuracy is :' ,round(r2_score(y_test,y_pred_3)*100.00,2),'%')
```
